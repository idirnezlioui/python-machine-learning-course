## 2 Fonctions

En Python, une **fonction** est un bloc de code r√©utilisable con√ßu pour ex√©cuter une t√¢che sp√©cifique.  
Les fonctions permettent de **structurer le code**, d‚Äô√©viter les r√©p√©titions et de le rendre plus **lisible** et **modulaire**.

---

###  D√©finir une fonction

La meilleure fa√ßon de cr√©er une fonction est d‚Äôutiliser la structure suivante : **def**

### fonction avec Return
La commande *return* permet de renvoyer une valeur √† l‚Äôendroit o√π la fonction a √©t√© appel√©e.


```python
# en python si en veux ecrire un fonction f(x) = x2 en utilise la fonction lambda
f=lambda x: x**2
print(f(5))  #affiche 25

#cette fonction elle utilisable plusieur fois avec des valeurs diff√©rentes
print(f(10)) #affiche 100
print(f(2.5)) #affiche 6.25 

#en peux ecrire une fonction mathematique plus complexe
g=lambda x,y: x**2 + y
print(g(2,3)) #affiche 7



#mais pour ecrire des fonction en utilise pas lambda  car il nous permet juste d'ecrire des fonctions math√©matiques simples
#mais nou en cherche a develloper une fonctionnalite cad une fonction qui nous permet de faire qeulqeue chose 
#pour √ßa en utilise le mot clef def
def e_potentiel(masse, hauteur,g):
    E=masse*hauteur*g #c'est une variable locale a la fonction (interne)
    print("L'√©nergie potentielle est de :", E, "Joules")

    #pour utilise la valeur de E en dehors de la fonction on utilise le mot clef return
    return E #elle permeet a la fonction de produire un resultat

#pour utiliser cette fonction on l'appelle avec des valeurs
e_potentiel(10,5,9.81)

#si on veux recuperer la valeur de retour on l'affecte a une variable
Energie=e_potentiel(2,3,9.81)
print("Energie potentielle r√©cup√©r√©e :", Energie, "Joules")
```
