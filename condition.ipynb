# Les Bases de Python - Structures de Contrôle

## 3. Structures de Contrôle

Dans le monde de la programmation, il existe trois principales structures de contrôle permettant de construire des algorithmes logiques et dynamiques :

1. Les alternatives `if / else`
2. Les boucles `for`
3. Les boucles `while`

Ces structures permettent au programme de prendre des décisions, répéter des actions ou exécuter du code de manière conditionnelle.

---

## 1. Alternatives `if / else`

La structure `if / else` permet de tester une ou plusieurs conditions et d’exécuter du code selon le résultat.

Une condition est considérée comme vraie uniquement si elle renvoie le booléen `True`.

### Syntaxe

```python
if condition1:
    # instructions si condition1 est vraie
elif condition2:
    # instructions si condition1 est fausse mais condition2 vraie
else:
    # instructions si aucune condition n'est vraie

# Les Bases de Python - Boucles 

## 4. Boucles

Les boucles permettent de répéter l'exécution d'un bloc de code plusieurs fois.  
Python propose principalement deux types de boucles : **`for`** et **`while`**.

---

## 1. Boucle `for`

Une boucle `for` permet de créer des algorithmes itératifs, c'est-à-dire qui effectuent une certaine tâche plusieurs fois de suite.  
La boucle parcourt tous les éléments d’un objet **itérable**, tel qu'une liste, un dictionnaire, un `range`, un tableau NumPy, etc.

### Syntaxe

```python
for element in iterable:
    # instructions à répéter

**Exercice** : Suite de Fibonacci

Objectif : Implémentez la suite de Fibonacci [0, 1, 1, 2, 3, 5, 8, 13, 21, ...] à partir de deux nombres initiaux a = 0 et b = 1.
Chaque nombre suivant est obtenu en additionnant les deux nombres précédents.

Indices

Vous aurez besoin d'une boucle while.

Affichez la suite jusqu'à atteindre un nombre n choisi.

Il est possible de mettre à jour deux variables simultanément :
