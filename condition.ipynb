# Les Bases de Python - Structures de Contrôle

## 3. Structures de Contrôle

Dans le monde de la programmation, il existe trois principales structures de contrôle permettant de construire des algorithmes logiques et dynamiques :

1. Les alternatives `if / else`
2. Les boucles `for`
3. Les boucles `while`

Ces structures permettent au programme de prendre des décisions, répéter des actions ou exécuter du code de manière conditionnelle.

---

## 1. Alternatives `if / else`

La structure `if / else` permet de tester une ou plusieurs conditions et d’exécuter du code selon le résultat.

Une condition est considérée comme vraie uniquement si elle renvoie le booléen `True`.

### Syntaxe

```python
if condition1:
    # instructions si condition1 est vraie
elif condition2:
    # instructions si condition1 est fausse mais condition2 vraie
else:
    # instructions si aucune condition n'est vraie

````python
# la premieir structure de controle: la conditionnelle if/else/elif
#elle nou permet de cree des algorithme ou certain lignes de code ne sont execute que si une condition est verifiee

#exemple:
x=1
y=2

if x>0:
    print("x est positif")

print("\n ")
#si x est négatif et en execute rien ne vas se passe
if x>0:
    print(x," est positif")
else:
    print(x," est negatif")

#et generalement on ne fais pas des choix binaire, on a souvent plusieurs cas a traiter

if x>0:
    print(x," est positif") #elle nous indique que cett ligne est execute dans la seule condition si x>0
elif x==0:
    print(x," est nul")
else:
    print(x," est negatif")

#rappel: x>0,x==0 : sont des expressions booleennes qui renvoient True ou False(des booleens)
x>0 

#et la maintenat en peux utilise des operateur logique pour combiner des expressions booleennes
#et : and
#ou : or
#non : not
if (x>0) and (y>0):
    print("x et y sont positifs")   



````

````python
#on peux aussi utilise la fonction sign dans une boucle for
for i in range(10,-10,-1):
    sign(i)
````

## La boucle `while`

Pour finir, nous allons parler du troisième type de structure de contrôle : **la boucle `while`**.


````python
#cette boucle la en peux dire que c'est un melange entre la boucle for et la conditionnelle if/elif/else
#puisque a chaque iteration de la boucle for on appelle la fonction sign qui elle meme

x=0
while x<10: 
    print(x)
    x+=1

#ce code il vas imprimer las valeur de x puis il vax faire x+1 tant que x est inferieur a 10
````
